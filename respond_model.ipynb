```python
# Load data from JSON file
with open('data/intent_data.json') as file:
    data = json.load(file)

# Extract intents, statements, and responses
intents = []
statements = []
responses = []

for item in data:
    intent = item['intent']
    statement = item['statement']
    
    if item.get("responses") != None:
        response_list = item["responses"]
        for response in response_list:
            intents.append(intent)
            statements.append(statement)
            responses.append(response)
    else:
        response = item["response"]
        intents.append(intent)
        statements.append(statement)
        responses.append(response)


# Verify the extracted data
print(intents)
print(statements)
print(responses)
```
